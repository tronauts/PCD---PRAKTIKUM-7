Proses Filter Rerata Pada Citra Input 1

```python
import numpy as np

# Mengimpor modul numpy untuk pengolahan array dan matriks

# Inisialisasi m1 dan n1 sesuai dengan ukuran citra
m1 = 10
n1 = 10

# Inisialisasi array citra dan output
copyCitra1 = np.zeros((m1, n1))  # Ganti dengan citra yang sesuai
output1 = np.zeros((m1, n1))

# Iterasi baris dari 1 hingga m1-1
for baris in range(1, m1-1):
    # Iterasi kolom dari 1 hingga n1-1
    for kolom in range(1, n1-1):
        # Mendapatkan posisi baris dan kolom saat ini
        a1 = baris
        b1 = kolom
        # Menghitung jumlah dari nilai sekitar pixel saat ini
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1+1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]
        # Mengupdate nilai output1 dengan rata-rata dari nilai sekitar
        output1[a1, b1] = (1/9 * jumlah)
```

```python
import numpy as np

# menginisialisasi m2 dan n2 sebelumnya
m2 = 10
n2 = 10

# Inisialisasi array copyCitra2 dan output2 (contoh dummy)
copyCitra2 = np.zeros((m2, n2))
output2 = np.zeros((m2, n2))

# Looping untuk setiap baris dalam range 1 sampai m2-1
for baris1 in range(1, m2-1):
    # Looping untuk setiap kolom dalam range 1 sampai n2-1
    for kolom1 in range(1, n2-1):
        # Mendefinisikan nilai a1 dan b1 sebagai baris1 dan kolom1
        a1 = baris1
        b1 = kolom1
        
        # Menghitung jumlah dari nilai sekeliling pada copyCitra2
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1+1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1]
        
        # Menghitung nilai output2 berdasarkan jumlah sekeliling
        output2[a1, b1] = (1/9) * jumlah
```
