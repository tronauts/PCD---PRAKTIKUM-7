```python
#ARDIANSYAH 1207070018
#TEKNIK ELEKTRO
#PENGOLAHAN CITRA DIGITAL

#IMPORT LIBRARY
import cv2  # Mengimpor modul cv2 untuk pemrosesan gambar
import numpy as np  # Mengimpor modul numpy untuk operasi matriks
from skimage import data  # Mengimpor modul data dari skimage untuk mengambil gambar contoh
import matplotlib.pyplot as plt  # Mengimpor modul matplotlib untuk membuat visualisasi
%matplotlib inline  
```

```python
img = data.camera()  # Mengambil gambar contoh 'camera' dari modul data

row, column = img.shape  # Mengambil ukuran baris dan kolom gambar

img1 = np.zeros((row,column),dtype = 'uint8')  # Membuat matriks nol dengan ukuran yang sama dengan gambar sebagai citra output

min_range = 10  # Menentukan batas bawah rentang piksel yang akan diubah menjadi putih
max_range = 60  # Menentukan batas atas rentang piksel yang akan diubah menjadi putih

# Melakukan iterasi untuk setiap piksel pada gambar
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:  # Jika piksel berada dalam rentang yang ditentukan
            img1[i,j] = 255  # Setel piksel di citra output menjadi putih
        else:
            img1[i,j] = 0  # Setel piksel di citra output menjadi hitam
```

PLOT IMAGE

```python
# Membuat plot dengan 2 baris dan 2 kolom untuk menampilkan gambar dan histogramnya
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

# Menampilkan citra input pada sumbu pertama
ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")

# Menampilkan histogram citra input pada sumbu kedua
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

# Menampilkan citra output pada sumbu ketiga
ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")

# Menampilkan histogram citra output pada sumbu keempat
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
